```python
class_dict = {'blues' : 0,
              'classical': 1,
              'country': 2,
              'disco' : 3,
              'hiphop' : 4,
              'jazz' : 5,
              'metal' : 6,
              'pop' : 7,
              'reggae': 8,
              'rock' : 9    
}
```

```python
class_dict.get('blues')
```

```python
def collect_data():
    data = []
    # Collect all images from the 10 genres
    for genre in os.listdir("../Data/images_original"):
        for image in os.listdir("../Data/images_original/" + genre):
            image_path = "../Data/images_original/" + genre + "/" + image
            grayscale_img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
            grayscale_img = grayscale_img[35:-36:, 54:-43]
            grayscale_img = torch.tensor(grayscale_img)
            data.append((grayscale_img.reshape(1,217,335), torch.tensor(class_dict.get(genre)).reshape(1)))
            
    return data           
```

```python
collect_data()[0][0].shape
```

```python
data = collect_data()
```

```python
len(data)
```

```python
from torch.utils.data import DataLoader
```

```python
dataloader = DataLoader(data[:10], batch_size=1, shuffle=True)
for i, batch in enumerate(dataloader):
    print(i, batch)
```

```python
import random
data = random.shuffle(data)
```
